```python
import geopandas
import pandas as pd
import numpy as np
import random
import math
from math import sqrt
import matplotlib.pyplot as plt
from shapely.geometry import Point
import folium
```

```python
resfil_raw = pd.read_csv('data/RVU_resfil.csv')
resfil_raw = resfil_raw[resfil_raw['dagtyp'] == 1]
```

```python
kommun = geopandas.read_file('data/kommuns.geojson')
kommun = kommun.to_crs('EPSG:4326')
```

```python
kommun['kommun'] = kommun.kommun.astype(int)
kommun = kommun.set_index('kommun')
```

```python
i = resfil_raw[['rf1_komkod', 'rf4_komkod']].replace({99.0: np.nan, 99: np.nan}).dropna().index
resfil_raw = resfil_raw.loc[i]
```

```python
resfil_raw['rf1_komkod'] = resfil_raw.rf1_komkod.astype(int)
resfil_raw['rf4_komkod'] = resfil_raw.rf4_komkod.astype(int)
```

```python
dd = {}
for person in resfil_raw.itertuples():
    skomm = person.rf1_komkod
    fkomm = person.rf4_komkod
    value = person.individvikt
    if (skomm,fkomm) in dd:
        dd[(skomm,fkomm)] += value
    elif (fkomm,skomm) in dd:
        dd[(fkomm,skomm)] += value
    else:
        dd[(skomm,fkomm)] = value
```

```python
summ = sum([d for d in dd.values()])
```

```python
c=500
```

```python
COORD = [55.63135, 13.70622]
m = folium.Map(location=COORD, zoom_start=8)

for k,tr in dd.items():
    if -1 in k:
        continue

    try:
        s = kommun.loc[k[0]].geometry.centroid
        e = kommun.loc[k[1]].geometry.centroid
    except KeyError as e:
        print(e)
        continue
        
    if s == e:
        folium.Circle((s.coords[0][1], s.coords[0][0]), weight=tr/summ*c, tooltip=str(tr)).add_to(m)
    else:
        folium.PolyLine([(x.coords[0][1], x.coords[0][0]) for x in [s,e]], weight=tr/summ*c, tooltip=str(tr)).add_to(m)
m.save('data/demand-rvu.html')
```
